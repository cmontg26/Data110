```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
plt.style.use("ggplot")  # Sets the style for the plots
```

```python
# Enter crime data
x = ['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft','Murder','Rape','Robbery'] # types of crimes
y = [264.1,250.7,1347.2,318.7,5.7,38.0,66.5] # crime statistics

# Set x-axis tick label size
ax=plt.gca()
ax.tick_params(axis='x', labelsize=6)

# Change size of columns to fit graph
plt.bar(x,y, width=0.25, label = '2023 Offenses', color= ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd','#1f77b4', '#ff7f0e'])

# Add titles and labels
plt.title ('Offenses per 100K People (2023)')
plt.xlabel('Crime Types')
plt.ylabel('Crime Rates (per 100,000)')

plt.legend()
plt.show()
```

```python
# Enter crime data
data = {'Crime Types': ['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft','Murder','Rape','Robbery'], 'Crime Stats':[-3.3,-8.1,-4.9, 12.0, -12.0, -9.8, -0.8]}

# Create the DataFrame
df=pd.DataFrame(data)

#Plot the horizontal bar chart
plt.style.use('bmh')
plt.figure(figsize=(10,6))
plt.barh(df['Crime Types'], df['Crime Stats'], color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd','#1f77b4', '#ff7f0e'])

# Set y-axis tick label size
ax=plt.gca()
ax.tick_params(axis='y', labelsize=10)
plt.grid(axis='x', linestyle='-', color='gray')

# Add titles and labels
plt.title ('Crime Rate % Changes (2022 v 2023)')
plt.xlabel('Crime Rates (per 100,000)')
plt.ylabel('Crime Types')


plt.show()
```

```python
# Optional Task: Combined Bar Chart for Crime Rates in 2022 and 2023

from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

# Enter crime data
x = ['Aggravated Assault','Burglary','Larceny','Motor Vehicle Theft','Murder','Rape','Robbery'] # types of crimes
y = [264.1,250.7,1347.2,318.7,5.7,38.0,66.5] # 2023 crime stats
z = [255.4,230.4,1281.2,356.9,5.0,34.3,66.0] # 2022 crime stats

# Set x-axis tick label size
ax=plt.gca()
ax.tick_params(axis='x', labelsize=6)

# Change size of columns to fit graph
plt.bar(x,y, width=0.4, label = '2023 Offenses')
plt.bar(x,z, width=0.4, label= '2022 Offenses', bottom=y)

# Add titles and labels
plt.title ('Offenses per 100K People (2023)')
plt.xlabel('Crime Types')
plt.ylabel('Crime Rates (per 100,000)')

plt.legend()
plt.show()
```
