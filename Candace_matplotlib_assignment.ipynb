```python
import matplotlib.pyplot as plt
print(plt.style.available)
```

```python
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
import numpy as np

# Define the range for x
x = np.linspace(-2, 4, 100)

# Calculate the f(x)
y=x**2-2*x+1

# Create the plot
plt.plot(x,y)

# Add labels and title for clarity
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Basic Line Plot")
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Define the range for x
x = np.linspace(0.1, 10, 200)

# Calculate the f(x)
y1=x**3
y2=np.exp(x)
y3=np.log(x) #(for x>0)

# Create the plot
plt.plot(x,y1, label='y=x^3', color='blue')
plt.plot(x,y2, label='y=e^x', color='red')
plt.plot(x,y3, label='y=log(x)', color='green')

# Add legend for clarity
plt.legend()

# Add labels and title for clarity
plt.xlabel("x")
plt.ylabel("y")
plt.title("Multiple Lines")
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Define the range for x
x = np.linspace(0.1, 10, 200)

# Calculate the f(x)
y1=x**2
y2=x**3
y3=x**4
y4=x**5

# Create the plot
plt.plot(x,y1, label='y=x^2', color='blue',linestyle='dotted')
plt.plot(x,y2, label='y=x^3', color='red', linestyle='dashdot')
plt.plot(x,y3, label='y=x^4', color='green', linestyle=':')
plt.plot(x,y4, label='y=x^5', color='orange', linestyle='--')

# Add legend for clarity
plt.legend()

# Add labels and title for clarity
plt.xlabel("x")
plt.ylabel("y")
plt.title("Styling Lines")
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Define the range for x
x = np.linspace(-10, 10, 200)

# Calculate the f(x)
y=1/(x**2+1)

# Create the plot
plt.plot(x,y, label=r'$f(x)=\frac{1}')

#Set custom x and y limits
plt.xlim(-5,5)
plt.ylim(0,1.1)
plt.axis('tight')

# Add labels and title for clarity
plt.xlabel("x")
plt.ylabel("y")
plt.title("Axes Limits and Scaling with tight axis")
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Define the range for x
x = np.linspace(-2*np.pi, 2*np.pi, 400)

# Calculate the f(x)
y= x* np.cos(x)

# Create scatter points at pi/4 intervals
x_scat= np.arange(-2*np.pi, 2*np.pi, np.pi/4)
y_scat= x_scat*np.cos(x_scat)

# Create the plot
plt.figure(figsize=(8,6))
plt.scatter(x_scat, y_scat, marker= 'o', color= 'blue', s= 36, facecolors= 'red', edgecolors= 'black', linewidths= 1.5, label='Scatter Points')
plt.plot(x, y, color='blue', linestyle='dotted', label='y = x cos(x)')

# Add labels and title for clarity
plt.xlabel("x")
plt.ylabel("y")
plt.title("Combined Plot")
plt.grid(True)
plt.legend()

# Show the plot
plt.show()
plt.style.use('seaborn-v0_8-whitegrid')
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range for x
x = np.linspace(0, 4*np.pi, 200)

# Calculate the f(x)
y1= x* np.sin(x)
y2= x* np.sin(x)
y3= x* np.sin(x)

# Create subplot function with number of rows and columns
figure, axis = plt.subplots(1,3, figsize=(15, 5))

# Plot for original plot with dark_background style
# Using "with" command for individual plots
with plt.style.context('dark_background'):
    axis[0].plot(x,y1, marker= 'o', color= 'green', linewidth= 1.5)
    axis[0].set_title("Original Plot with dark_background")

# Plot for seaborn-v0_8-whitegrid style
# have to add white background
with plt.style.context('seaborn-v0_8-whitegrid'):
    axis[1].plot(x,y2, linestyle= '--', color= 'blue', linewidth= 1.5)
    axis[1].set_title("seaborn-v0_8-whitegrid")
    axis[1].set_facecolor('white')

# Plot for Alternative Style #3 with fivethirtyeight style
with plt.style.context('fivethirtyeight'):
    axis[2].plot(x,y3)
    axis[2].set_title("fivethirtyeight")

# Add labels and title for clarity
figure.suptitle("Style Comparison")

# Adjust layout to prevent titles overlapping
plt.tight_layout(rect=[0, 0.03, 1, 0.95])


# Show the plot
plt.show()
```
