```python
df=sns.load_dataset('titanic')
df.describe()
```

I will use the variables fare, sex, pclass, age, and survived to interpret socioeconomic status of passengers. I will also compare fare across gender and passenger class (pclass) variables to highlight any differences across passenger classes and gender

```python
#Call out required columns
required_cols = ['fare', 'sex', 'pclass', 'age', 'survived']

#Call out numeric columns
num_cols = ['fare', 'age']

#Call out categorical columns
cat_cols = ['sex', 'pclass', 'survived']
```

```python
sns.histplot(df['fare'],kde=True, bins=15, color = 'green', edgecolor= 'black')
plt.style.use('ggplot')
plt.xlabel("Fare of Titanic Passengers")
plt.xticks([0,100, 200, 300, 400, 500])
plt.ylabel("Quantity of Survivors")
plt.grid(axis='x')
plt.title("Titanic Survivor Fares")
plt.show()
```

```python
df.dropna(subset=['fare'], inplace=True)

plt.figure(figsize=(10, 6))

# Plot KDEs
sns.kdeplot(data=df, x='fare', hue='survived', fill=True, common_norm=False, alpha=0.5)

plt.title('Fare of Survived vs. Passed Passengers', fontsize=16)
plt.xlabel('Fare', fontsize=12)
plt.ylabel('Probabability of Not Surviving', fontsize=12)
plt.legend(title='Survival', labels=['Passed', 'Survived'])
#Limit x-axis for better visibility
plt.xlim(0, 300)
plt.show()
```

#Candace's Interpretation

The shapes reveals both groups were dominated by low-fare passengers, however the survival distribution is less skewed and more spread out for higher fares, indicating a strong correlation between paying a higher fare  and surviving the big accident.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='pclass', y='fare', hue='sex', data=df)
plt.title('Fare by Passenger Class and Sex')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')
plt.legend(title='Gender')
plt.show()
```

#Candace's Interpretation

There is not much difference between second and third class passengers regarding gender. However more women were first class compared to males.


```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='sex', y='age', hue='sex', data=df)
plt.title('Age Ranges by Gender')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

#Candace's Interpretation

The primary age range of most passengers for both gender is between 20 and 40 years old.

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='sex', y='age', data=df, inner=None, color="skyblue")
# Strip plot for individual observations (jitter added to reduce overlap)
sns.stripplot(x='sex', y='age', data=df, jitter=True, size=4, alpha=0.6, color='red')
plt.title('Age Range by Gender with Individual Observations')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

#Candace's Interpretation

The oldest passenger is 80 years old male and there were quite a few infants onboard.
