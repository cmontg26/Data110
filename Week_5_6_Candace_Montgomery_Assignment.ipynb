```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/Reben80/Data110-21843--fall25/refs/heads/main/dataset/titanic.csv')
df.describe()
```

```python
sns.histplot(df['age'],kde=True, bins=30)
plt.style.use('bmh')
plt.xlabel("Age of Titanic Passengers")
plt.xticks([0,20,40,60,80])
plt.ylabel("Quantity of Survivors")
plt.title("Titanic Survival Age Demographics")
plt.show()
```

```python
sns.catplot(data=df, x="age", y="sex", hue="sex", kind="violin")
plt.xlabel("Age of Titanic Passengers")
plt.ylabel("Gender of Titantic Passengers")
plt.title("Titantic Demographics")
plt.show()
```

```python
sns.catplot(data=df, x="survived", y="age", hue="sex", kind="box")
plt.xlabel("Survival of Titanic Passengers")
plt.ylabel("Quantity of Titantic Passengers")
plt.title("Titantic Demographics")
plt.show()
```
