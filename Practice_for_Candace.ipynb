```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

days = np.arange(1, 31)
prices = 100 + np.cumsum(np.random.randn(30))

plt.plot(days, prices, marker='o', linestyle='-', color='green')
plt.title("Simulated Stock Price")
plt.xlabel("Day")
plt.ylabel("Price ($)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 1. Prepare your data
# For demonstration, let's create a sample correlation matrix
data = np.random.rand(10, 10) # Generate random data
df = pd.DataFrame(data, columns=[f'Col_{i}' for i in range(10)],
                  index=[f'Row_{i}' for i in range(10)])

# df = sns.load_dataset("flights").pivot_table(index="month", columns="year", values="passengers")
# correlation_matrix = df.corr()

# 2. Create the heatmap
plt.figure(figsize=(8, 6)) # Adjust figure size as needed
sns.heatmap(df, annot=True, cmap='viridis', fmt=".2f", linewidths=.5)
# - annot=True displays the data values on the heatmap
# - cmap sets the color map (e.g., 'viridis', 'YlGnBu', 'coolwarm')
# - fmt=".2f" formats the annotations to two decimal places
# - linewidths adds lines between cells

# 3. Add titles and labels (optional)
plt.title('Sample Heatmap')
plt.xlabel('Columns')
plt.ylabel('Rows')

# 4. Display the heatmap
plt.show()
```
