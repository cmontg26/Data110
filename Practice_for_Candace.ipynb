```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

days = np.arange(1, 31)
prices = 100 + np.cumsum(np.random.randn(30))

plt.plot(days, prices, marker='o', linestyle='-', color='green')
plt.title("Simulated Stock Price")
plt.xlabel("Day")
plt.ylabel("Price ($)")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Prepare the data
# Create a correlation matrix
data = np.random.rand(10, 10) # Generate random data
df = pd.DataFrame(data, columns=[f'Col_{i}' for i in range(10)],
                  index=[f'Row_{i}' for i in range(10)])

# df = sns.load_dataset("flights").pivot_table(index="month", columns="year", values="passengers")
# correlation_matrix = df.corr()

# 2. Create the heatmap
plt.figure(figsize=(8, 6)) # Adjust figure size as needed
sns.heatmap(df, annot=True, cmap='inferno', fmt=".2f", linewidths=.5)
# - annot=True displays the data values on the heatmap
# - cmap sets the color map
# - fmt=".2f" formats the annotations to two decimal places
# - linewidths adds lines between cells

# 3. Add titles and labels
plt.title('Heatmap for Practice')
plt.xlabel('Columns')
plt.ylabel('Rows')

# 4. Display the heatmap
plt.show()
```

```python
# Made-up data
data = {'Low Income': 30, 'Low Food Access': 15, 'LILA': 5}

# Create a pictogram chart
fig = plt.figure(
    FigureClass=Waffle,
    rows=5,  # Number of rows in the waffle chart
    values=data,
    colors=["#4C8CB5", "#B7CBD7", "#C0C0C0"], # Colors for each category
    icons='user',  # Icon to use for the pictogram (e.g., 'star', 'user', 'heart' from Font Awesome)
    font_size=18,  # Size of the icons
    icon_legend=True, # Display a legend with icons
    legend={'labels': [f"{k}: {v}" for k, v in data.items()], 'loc': 'upper right', 'bbox_to_anchor': (0, -0.2)})

plt.title("Gradient Person Chart", fontsize=14)
plt.show()
```
