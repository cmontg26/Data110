##Make a new pairplot for the penguins dataset with these settings:

###**Use a different column for hue than before (for example, island or sex).**

* Choose different markers such as ["o", "s", "D"].
* Pick a different color palette (for example, "magma" or "coolwarm").
* Keep the diagonal as KDE (diag_kind="kde").

```python
# Columns to use
num_cols = ["bill_length_mm", "bill_depth_mm", "flipper_length_mm", "body_mass_g"]
hue_col = "island"  # different from species; works well with 3 markers
cols_used = num_cols + [hue_col]

# Drop missing values for the columns used
df = penguins[cols_used].dropna(subset=cols_used)

# Optional: fix category order for consistent marker/color assignment
island_order = ["Biscoe", "Dream", "Torgersen"]

# Pairplot: different hue, custom markers, palette, and KDE diagonals
g = sns.pairplot(
    df,
    vars=num_cols,
    hue=hue_col,
    hue_order=island_order,
    diag_kind="kde",
    markers=["o", "s", "D"],
    palette="coolwarm",
    plot_kws={"alpha": 0.8, "s": 40},
    diag_kws={"fill": True, "common_norm": False}
)

g.fig.suptitle("Penguins Pairplot  hue=island, markers=['o','s','D'], palette='coolwarm'", y=1.02)
plt.show()
```

#Task 5: Create a 22 Grid with Mixed Plot Types
Make a figure with four subplots arranged in a 22 grid using
fig, axes = plt.subplots(2, 2, figsize=(10, 8)).

**Use the penguins dataset and include a mix of plot types:**

* (Top-left): sns.barplot()  show body mass by species.
* (Top-right): sns.scatterplot()  plot bill length vs bill depth, color by species.
* (Bottom-left): sns.histplot()  distribution of flipper length, separated by species (hue="species").
* (Bottom-right): sns.boxplot()  compare body mass by island.
* Add a title for each subplot, and call plt.tight_layout() at the end.

**Deliverable**: one combined figure (saved as .png, around 150 dpi) that clearly shows all four plots with appropriate labels and colors.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
penguins = sns.load_dataset("penguins")

# Create a 22 grid
fig, axes = plt.subplots(2, 2, figsize=(10, 8))

# --- Top-left: Barplot (body mass by species) ---
sns.barplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    hue="species",
    legend = False,
    ax=axes[0, 0],
    palette="viridis"
)
axes[0, 0].set_title("Average Body Mass by Species")
axes[0, 0].set_xlabel("Species")
axes[0, 0].set_ylabel("Body Mass (g)")

# --- Top-right: Scatterplot (bill length vs bill depth, color by species) ---
sns.scatterplot(
    data=penguins,
    x="bill_length_mm",
    y="bill_depth_mm",
    hue="species",
    ax=axes[0, 1],
    palette="coolwarm"
)
axes[0, 1].set_title("Bill Length vs Bill Depth by Species")
axes[0, 1].set_xlabel("Bill Length (mm)")
axes[0, 1].set_ylabel("Bill Depth (mm)")

# --- Bottom-left: Histplot (flipper length by species) ---
sns.histplot(
    data=penguins,
    x="flipper_length_mm",
    hue="species",
    multiple="stack",
    ax=axes[1, 0],
    palette="magma"
)
axes[1, 0].set_title("Distribution of Flipper Length by Species")
axes[1, 0].set_xlabel("Flipper Length (mm)")
axes[1, 0].set_ylabel("Count")

# --- Bottom-right: Boxplot (body mass by island) ---
sns.boxplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    hue = "species",
    legend = False,
    ax=axes[1, 1],
    palette="viridis"
)
axes[1, 1].set_title("Body Mass by Island")
axes[1, 1].set_xlabel("Islands")
axes[1, 1].set_ylabel("Body Mass (g)")

# Adjust layout and save figure
plt.tight_layout()
plt.show()
```
